```python
from turtle import title
import matplotlib.pyplot as plt
import numpy as np

# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2,ax3) = plt.subplots(1, 3, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, )
axs[0, 0].plot(x, y)
axs[0,1].plot(x,y)
axs[1, 1].scatter(x, y)

# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')

# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)

# Create figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax = plt.subplots(num=10, clear=True)
```

```python
a = np.array([50,40,30,20,10,60])
b = np.array([20,30,10,50,40,25])
# color  = ['red','blue','green','black','yellow']
color = [70.5,20,40,10,100,80]
size = [60,30,100,10,60,45]
area1= np.ma.masked_where(a<=40,a)
area2= np.ma.masked_where(a>40,a)
```

how to use color as an array is that whenever u want to show a third feature on the graph w.r.t some other scaling, then u can have color as that feature(numeric) and hence correspopnding to each datapoint we can have a specific color according to the third feature

```python

plt.scatter(a,b,c = color,s = area1,cmap = 'Accent',marker= 's')
plt.scatter(a,b,c = color,s = area2,cmap = 'Accent',marker= '^')
#colored based on 'c'
plt.colorbar()
```

```python
plt.scatter(a,b,c = area1,s = area1,cmap = 'Accent',marker= 's')
plt.scatter(a,b,c = area2,s = area2,cmap = 'Accent',marker= '^')
#colored based on 'c'
# here the cmap adjust according to c 
# so we see a change in scale in colorbar.
plt.colorbar()
plt.show()
```

default cmap: 'viridis'

markers: 's':square; 'o'(default): circle; '^': triangle etc.
